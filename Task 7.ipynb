# Data Wrangling

```python
import pandas as pd
import numpy as np
```

```python
# Handling missing values in a dataset.

income_data = pd.read_csv(r'C:\Users\mahmud kamai\Downloads/income_data.csv')
```

```python
income_data.head()
```

```python
income_data.describe()
```

```python
income_data.info()
# in this case missing values are present in Age, Income and Gender.
```

```python
# Identifying the column that contains missing values.
missing_values = income_data.isnull().sum()
missing_values
```

```python
#calculate the percentage of missing data in each column.
percentage_of_missing_values = (missing_values / len(income_data)) * 100
percentage_of_missing_values
```

```python
missing_data = pd.DataFrame({'missing values' : missing_values, 'percentage' : percentage_of_missing_values})
missing_data
```

```python
# Numerical Columns 
# filling missing values in numerical columns like age and income using statistical method.

income_data['Age'].fillna(income_data['Age'].mean(), inplace=True)
income_data['Income'].fillna(income_data['Income'].mean(), inplace=True)
```

```python
income_data.head()
```

```python
income_data.sample(20)
```

```python
# Fill missing values in categorical columns like 'Gender' with the mode. 

income_data['Gender'].fillna(income_data['Gender'].mode()[0], inplace=True)
income_data.tail()
```

```python
income_data.info()
```

```python
income_data.dropna()
```

# Handling Outliers

```python
import matplotlib
from matplotlib import pyplot as plt

# scatter plot

plt.scatter(income_data['Income'], income_data['Score'])
plt.title('Scatter plot of Income and Score')
plt.xlabel('Income')
plt.ylabel('Score')
plt.show
```

```python
income_data.describe()
```

```python
# Use statistical methods like the Interquartile Range (IQR) to identify and remove outliers in income.
# for income
Q1_income = income_data.Income.quantile(0.25)
Q3_income = income_data.Income.quantile(0.75)
IQR_income = Q3_income - Q1_income

print(Q1_income)
print(Q3_income)
print(IQR_income)
```

```python
# find the lower limit and upper limit

income_up = Q3 + 1.5*IQR
income_lo = Q1 - 1.5*IQR
print(income_up)
print(income_lo)
```

```python
outliers_income = income_data[(income_data['Income'] > income_up) | (income_data['Income'] < income_lo)]
print(outliers_income)
```

```python
# Use statistical methods like the Interquartile Range (IQR) to identify and remove outliers in score.
# for income
Q1_score = income_data.Score.quantile(0.25)
Q3_score = income_data.Score.quantile(0.75)
IQR_score = Q3_score - Q1_score

print(Q1_score)
print(Q3_score)
print(IQR_score)
```

```python
# find the lower limit and upper limit

score_upper_limit = Q3_score + 1.5*IQR_score
score_lower_limit = Q1_score - 1.5*IQR_score
print(score_upper_limit)
print(score_lower_limit)
```

```python
outliers_score = income_data[(income_data['Score'] > score_upper_limit) | (income_data['Score'] < score_lower_limit)]
print(outliers_score)

# note: this shows that there is no outliers in the score column.
```

```python
cleaned_income_data = income_data[(income_data['Income'] > income_lo) & (income_data['Income'] < income_up)]
cleaned_income_data
```
